A) Expected value 

```python
import numpy as np
def exp_val(values,probabilities):
  return sum([v*p for v,p in zip(values,probabilities)])
print("Exected Value:",exp_val([2,3,4],[0.2,0.3,0.5]))
```

B) Mean

```python
n=np.arange(1,11)
print(n)
#np.mean gives you the mean of n elements 
print("Mean Value:",np.mean(n))
print("Average Value:",np.average(n))
```

c)standard deviation

```python
n=np.arange(1,10)
print(n)
#np.std gives you the standard deviation of an numpy array 
print("Standard Deviation:",np.std(n))
```

d)variance

```python
n=np.arange(0,10)
print(n)
#np.var gives you the variance of elements in a numpy array
print("Variance:",np.var(n))
```

c)covariance

```python
x=np.array([1,2,3,4])
#np.cov gives you the covariance of an numpy array
print("Covariance",np.cov(x))
```

f) covariance with two arrays \\

```python
x=np.array([0,1,2])
y=np.array([2,1,0])
print("Array one")
print(x)
print("\nArray two")
print(y)
print("\nCovariance using two matrix")
#we can also calculate the covariance of two arrays using np.cov
print(np.cov(x,y))
```
